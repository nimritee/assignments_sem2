```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('analysis.csv',names=["machine", "task", "time"])
```

```python
information ={}
for i in range(len(data)):
    machine = data.iloc[i,0]
    task = data.iloc[i,1]
    if task in information:
        information[task] += 1
    else:
        information[task]= 1
```

```python
plt.plot(information.keys(), information.values())
plt.xlabel("time")
plt.ylabel("task")
plt.title("Analyzing Time taken to complete a Task")
plt.show() 
```

```python
machine_information ={}
for i in range(len(data)):
    machine = data.iloc[i,0]
    if machine in machine_information:
        machine_information[machine] += 1
    else:
        machine_information[machine]= 1
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
machines = machine_information.keys()
time = machine_information.values()
ax.bar(machines,time)
ax.set_xticks(np.arange(len(machines)))
plt.ylim(21000,)
plt.xlabel("machine")
plt.ylabel("time")
plt.title("Analyzing Total Time spent on Machine")
plt.show()
```
